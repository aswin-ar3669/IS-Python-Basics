```python
import scipy.stats as st
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
A = np.array([ 68.7,75.4,70.9,79.1,78.2])
B = np.array([ 62.7, 68.5, 63.1, 62.2, 60.3])
C = np.array([ 55.9, 56.1, 57.3, 59.2, 50.1])
D = np.array([ 80.7,70.3,80.9,85.4,82.3])
```

```python
sns.kdeplot(A)
sns.kdeplot(B)
sns.kdeplot(C)
sns.kdeplot(D)
```

```python
st.f_oneway(A,B,C,D)
```

```python

nd = pd.DataFrame({'A':  np.array([ 68.7,75.4,70.9,79.1,78.2]) ,
'B' :np.array([ 62.7, 68.5, 63.1, 62.2, 60.3]),
'C' : np.array([ 55.9, 56.1, 57.3, 59.2, 50.1]),
'D' : np.array([ 80.7,70.3,80.9,85.4,82.3])})
```

```python

nd1 = pd.melt(nd)
nd1
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=nd1['value'],
                          groups=nd1['variable'],
                          alpha=0.05)

print(tukey)
```

```python
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(nd1['value'], nd1['variable'])
result = mc.tukeyhsd()

print(result)
```

```python

club1 = np.array([25, 35, 45])
club2 = np.array([10, 20, 30])
club3 = np.array([35, 60, 85])
```

```python
sns.kdeplot(club1)
sns.kdeplot(club2)
sns.kdeplot(club3)
```

```python
st.shapiro(club1), st.shapiro(club2), st.shapiro(club3)
```

```python
st.f_oneway(club1, club2, club3)
```

```python
st.levene(club1, club2, club3)
```
